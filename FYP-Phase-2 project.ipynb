**About Dataset:**

Student Performance & Behavior Dataset
This dataset is real data of 5,000 records collected from a private learning provider.
The dataset includes key attributes necessary for exploring patterns, correlations, and insights related to academic performance.

**Columns**:

*   Student_ID: Unique identifier for each student.
*   First_Name: Student’s first name.
*   Last_Name: Student’s last name.
*   Email: Contact email (can be anonymized).
*   Gender: Male, Female, Other.
*   Age: The age of the student.
*   Department: Student's department (e.g., CS, Engineering, Business).
*   Attendance (%): Attendance percentage (0-100%).
*   Midterm_Score: Midterm exam score (out of 100).
*   Final_Score: Final exam score (out of 100).
*   Assignments_Avg: Average score of all assignments (out of 100).
*   Quizzes_Avg: Average quiz scores (out of 100).
*   Participation_Score: Score based on class participation (0-10).
*   Projects_Score: Project evaluation score (out of 100).
*   Total_Score: Weighted sum of all grades.
*   Grade: Letter grade (A, B, C, D, F).
*   Study_Hours_per_Week: Average study hours per week.
*   Extracurricular_Activities: Whether the student participates in extracurriculars (Yes/No).
*   Internet_Access_at_Home: Does the student have access to the internet at home? (Yes/No).
*   Parent_Education_Level: Highest education level of parents (None, High School, Bachelor's, Master's, PhD).
*   Family_Income_Level: Low, Medium, High.
*   Stress_Level (1-10): Self-reported stress level (1: Low, 10: High).
*   Sleep_Hours_per_Night: Average hours of sleep per night.

The Attendance is not part of the Total_Score or has very minimal weight.


**Calculating the weighted sum**:

Total Score = a⋅Midterm+b⋅Final+c⋅Assignments+d⋅Quizzes+e⋅Participation+f⋅Projects

**Component	Weight (%)**


*   Midterm	15%
*   Final	25%
*   Assignments Avg	15%
*   Quizzes Avg	10%
*   Participation	5%
*   Projects Score	30%
*   Total	100%

## Train Decision Tree

### Subtask:
Train a Decision Tree Classifier on the training data (X_train, y_train).


**Reasoning**:
To train a Decision Tree Classifier, I need to first import the `DecisionTreeClassifier` class from `sklearn.tree`.



```python
from sklearn.tree import DecisionTreeClassifier
print("DecisionTreeClassifier imported successfully.")
```

**Reasoning**:
Now that the `DecisionTreeClassifier` is imported, I need to instantiate it and then fit it to the training data `X_train` and `y_train` to train the model.



```python
dt_classifier = DecisionTreeClassifier(random_state=42)
dt_classifier.fit(X_train, y_train)
print("Decision Tree Classifier trained successfully.")
```

**Changes after first splitting:**

The initial attempt to train the Decision Tree Classifier faced an issue where the target variable 'Total_Score' was continuous, while classification models require discrete labels. This was corrected by switching the target variable to 'Grade'.



```python
columns_to_drop_from_X = ['Student_ID', 'First_Name', 'Last_Name', 'Email', 'Grade', 'Total_Score']
X = df.drop(columns=columns_to_drop_from_X)
y = df['Grade']

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 78)

from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train.iloc[:,2:9] = sc.fit_transform(X_train.iloc[:,2:9])
X_test.iloc[:,2:9] = sc.transform(X_test.iloc[:,2:9])

from sklearn.tree import DecisionTreeClassifier
dt_classifier = DecisionTreeClassifier(random_state=42)
dt_classifier.fit(X_train, y_train)
print("Decision Tree Classifier trained successfully.")
```

## Evaluate Decision Tree

### Subtask:
Evaluate the trained Decision Tree Classifier using accuracy, precision, recall, and F1-score on the test data (X_test, y_test).


**Reasoning**:
To evaluate the trained Decision Tree Classifier, I need to first make predictions on the test set (`X_test`) using the `dt_classifier` model. Then, I will import the required evaluation metrics (accuracy, precision, recall, and F1-score) from `sklearn.metrics` to compare these predictions (`y_pred_dt`) against the actual target values (`y_test`) and print the results.



```python
y_pred_dt = dt_classifier.predict(X_test)

from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

accuracy_dt = accuracy_score(y_test, y_pred_dt)
precision_dt = precision_score(y_test, y_pred_dt, average='weighted', zero_division=0)
recall_dt = recall_score(y_test, y_pred_dt, average='weighted', zero_division=0)
f1_dt = f1_score(y_test, y_pred_dt, average='weighted', zero_division=0)

print(f"Decision Tree Classifier Metrics:")
print(f"  Accuracy: {accuracy_dt:.4f}")
print(f"  Precision: {precision_dt:.4f}")
print(f"  Recall: {recall_dt:.4f}")
print(f"  F1-Score: {f1_dt:.4f}")
```
